# World Cholera Dataset Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import math

```

```python
# Number of Cholera Cases and Deaths per Country
```

```python
Country_CholeraCases_Deaths=WCA.groupby("Country")[["Cholera Cases per 100,000 people", "Infant Mortality Rate (per 1,000 live births)"]].sum().reset_index()
```

```python
Country_CholeraCases_Deaths=Country_CholeraCases_Deaths.sort_values(by="Cholera Cases per 100,000 people",ascending=True)
print(Country_CholeraCases_Deaths)
```

```python
pd.DataFrame(Country_CholeraCases_Deaths)
```

```python
# Number of Cases and Deaths per Year
```

```python
Year_CholeraCases_Deaths=WCA.groupby("Year")[["Cholera Cases per 100,000 people", "Infant Mortality Rate (per 1,000 live births)"]].sum().reset_index()
print(Year_CholeraCases_Deaths)
```

```python
pd.DataFrame(Year_CholeraCases_Deaths)
```

```python
# Trend Over Time
```

```python
Yearly_trend=WCA.groupby("Year")["Cholera Cases per 100,000 people"].sum().reset_index()
print(Yearly_trend)
```

```python
pd.DataFrame(Yearly_trend)
```

```python
plt.figure(figsize=(10,6))
plt.plot(Yearly_trend["Year"],Yearly_trend["Cholera Cases per 100,000 people"])
plt.title("CASES BY YEAR")
plt.xlabel("Year")
plt.ylabel("Cholera Cases per 100,000 people")
plt.grid(False)
plt.show()
```

```python
# Cases/DeathRatio
```

```python
Total_Cholera=WCA["Cholera Cases per 100,000 people"].sum()
print(Total_Cholera)
```

```python
Total_Death=WCA["Infant Mortality Rate (per 1,000 live births)"].sum()
print(Total_Death)
```

```python
Cholera_Death_Ratio=Total_Cholera/Total_Death
print(Cholera_Death_Ratio)
```

```python
print(f"The Cholera to death ratio is 0.47 reprsenting 47% of the total mortality cases")
```

```python
# Top 10 Cholera Cases by Country
```

```python
Country_total=WCA.groupby("Country")["Cholera Cases per 100,000 people"].sum().reset_index()
```

```python
Top10Country=Country_total.sort_values(by="Cholera Cases per 100,000 people",ascending=False).head(10)
```

```python
colors=["orange","blue","green","yellow","black","red","violet","grey","brown","purple"]
plt.figure(figsize=(10,6))
sns.barplot(data=Top10Country,x="Country",y="Cholera Cases per 100,000 people",palette=colors)
plt.title("Top 10 Cholera Cases by Country")
plt.xlabel("Country")
plt.ylabel("Cholera Cases per 100,000 people")
plt.show()
```
